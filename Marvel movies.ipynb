## Maiores Bilheterias da Marvel

fonte: https://www.kaggle.com/datasets/davidgdong/marvel-cinematic-universe-box-office-dataset

- As maiores bilheterias da Marvel;
- Os maiores faturamentos da Marvel.
- Qual é a fase da Marvel que mais faturou com bilheteria nos cinemas?
- Qual a fase teve mais avaliação positiva?

```python
import pandas as pd
```

```python
dados = pd.read_csv("mcu_box_office.csv")
dados.head()
```

```python
dados.info()
```

```python
dados['release_date'] = pd.to_datetime(dados['release_date'], format = '%m/%d/%Y')
```

```python
# tentei o replace(',', '.') separando as casas decimais, mas esse formato não funcionou...

dados['production_budget'] = dados['production_budget'].str.replace(',','')

dados['opening_weekend'] = dados['opening_weekend'].str.replace(',','')

dados['domestic_box_office'] = dados['domestic_box_office'].str.replace(',','')

dados['worldwide_box_office'] = dados['worldwide_box_office'].str.replace(',','')
```

```python
dados.head()
```

### conversões

mudanças nescessarias:
- mudar o formato da data;<br>
- converter o formato dos números __object__ para __int64__.


```python
dados['production_budget'] = pd.to_numeric(dados['production_budget'])

dados['opening_weekend'] = pd.to_numeric(dados['opening_weekend'])

dados['domestic_box_office'] = pd.to_numeric(dados['domestic_box_office'])

dados['worldwide_box_office'] = pd.to_numeric(dados['worldwide_box_office'])
```

```python
dados['production_budget'] = dados['production_budget'].astype('int64')

dados['opening_weekend'] = dados['opening_weekend'].astype('int64')

dados['domestic_box_office'] = dados['domestic_box_office'].astype('int64')

dados['worldwide_box_office'] = dados['worldwide_box_office'].astype('int64')
```

```python
dados.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
preto = 'black'
```

```python
fig, ax = plt.subplots(figsize=(12,13))
sns.barplot(data=dados, y='movie_title', x='production_budget')
plt.style.use('ggplot')
plt.ylabel('Filme', fontsize=13)
plt.xlabel('Orçamento Total de Produção', fontsize=13)
plt.title('Orçamento Total de Produção de cada Filme - USD($)', fontsize=19)
ax.tick_params(labelsize=13, labelcolor=preto)
#ax.grid(False)
```

### bilheteria mundial

```python
fig, ax = plt.subplots(figsize=(12,13))
sns.barplot(data=dados, x='worldwide_box_office', y='movie_title')
plt.title('Receita Bilheteria mundial - USD($)', fontsize=19)
plt.xlabel('bilheteria', fontsize=13)
plt.ylabel('filme', fontsize=13)
ax.tick_params(labelsize=13, labelcolor=preto)
#ax.grid(False)
```

### Marvel fases

```python
dados['mcu_phase'].value_counts()
```

```python
dados.groupby('mcu_phase').agg({'worldwide_box_office':'sum'})
```
